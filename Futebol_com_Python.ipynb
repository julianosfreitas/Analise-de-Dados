Quando utilizamos as *compara√ß√µes* da biblioteca *pandas* dessa forma, precisa usar **&** para o operador **and** e **|** para operador **or**.

Como *filtrar* apenas as **vit√≥rias** dos **goleiros**? üß§‚úÖ
-> GOLEIROS **E** VIT√ìRIAS

```python
dados[(dados['Posi√ß√£o'] == 'Goleiro') & (dados['Vit√≥ria']==True)]
```

Como *filtrar* apenas os **goleiros** e **zagueiros**? üß§üõ°Ô∏è

```python
dados[(dados['Posi√ß√£o'] == 'Goleiro') | (dados['Posi√ß√£o']=='Zagueiro')]
```

Como ver os **dados nulos**? ‚ùå -> fun√ß√£o **ISNA ( )**

```python
dados.isna()
```

```python
dados.isna().sum()
```

 Como tirar esses **dados nulos**? ‚ùå **>** ‚ôªÔ∏è***(Limpeza de dados)***

```python
colunas = []
for coluna in dados.columns:
  if dados[coluna].isna().sum != 0:
    colunas.append(coluna)
dados[colunas]
```

```python
dados.fillna(0, inplace = True)
```
