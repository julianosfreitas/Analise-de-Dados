```python
def arredondar_personalizado(valor):
    # Verifica se o valor est√° entre 1 e 5
    if not (1.0 <= valor <= 5.0):
        raise ValueError("A avalia√ß√£o deve estar entre 1 e 5")

    # Lista de valores permitidos
    valores_permitidos = [1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]

    # Encontrar o valor mais pr√≥ximo na lista
    valor_arredondado = min(valores_permitidos, key=lambda x: abs(x - valor))

    return valor_arredondado

# Solicita as avalia√ß√µes ao usu√°rio
avaliacoes = {}
for categoria in ['fotografia', 'roteiro', 'atuacao', 'direcao', 'trilha sonora']:
    while True:
        try:
            # Pede a avalia√ß√£o e converte para float
            nota = float(input(f"Insira a nota para {categoria} (entre 1 e 5): "))
            # Verifica se a nota est√° no intervalo permitido
            if not (1.0 <= nota <= 5.0):
                raise ValueError
            avaliacoes[categoria] = nota
            break
        except ValueError:
            print("Valor inv√°lido. Insira um n√∫mero entre 1 e 5.")

# Calcular a m√©dia
media = sum(avaliacoes.values()) / len(avaliacoes)

# Arredondar a m√©dia e exibir
try:
    media_arredondada = arredondar_personalizado(media)
    print(f"Avalia√ß√£o: {media_arredondada}")
except ValueError as e:
    print(e)
```
